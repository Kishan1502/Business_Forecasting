```python
import pandas as pd
```

```python
make=pd.read_csv('machine_idle_time.csv')
make
```

```python
make['forecasted_idle_time']=make.idle_time.rolling(5).mean()
make
```

```python
make.loc[0]=[20,25.4]
make
```

```python
make=make.rename(columns={'idle_time': 'Peacock'})
make
```

```python
import matplotlib.pyplot as plt
```

```python
make.plot(y=['Peacock','forecasted_idle_time'])
plt.title("Peacock Vs. Forecaste sales")
plt.show()
```

```python
make['Yt-Ft'] = make['Peacock'] - make['forecasted_idle_time']
make
```

```python
make['|Yt-Ft|'] = make['Yt-Ft'].abs()
make
```

```python
make['|Yt-Ft|/Yt'] = make['|Yt-Ft|'] / make['Peacock']
make
```

```python
make.loc[:, '|Yt-Ft|'].mean()
```

```python
make.loc[:, '|Yt-Ft|/Yt'].mean()
```

```python
m= make['Yt-Ft']**2
m
```

```python
m.mean()#MSE
```

```python
import pandas as pd
```

```python
zoo=pd.read_csv('visitors_zoo.csv')
zoo
```

```python
zoo['forecasted_visitors']=zoo.visitors.rolling(2).mean()
zoo
```

```python
zoo.loc[0]=[25,46]
zoo
```

```python
zoo=zoo.rename(columns={'visitors': 'ZOO Visitors'})
zoo
```

```python
import matplotlib.pyplot as plt
```

```python
zoo.plot(y=['ZOO Visitors','forecasted_visitors'])
plt.title("Zoo Visitors Vs. Forecasted Visiotrs")
plt.show()
```

```python
zoo['Yt-Ft'] = zoo['ZOO Visitors'] - zoo['forecasted_visitors']
zoo
```

```python
zoo['|Yt-Ft|'] = zoo['Yt-Ft'].abs()
zoo
```

```python
zoo['|Yt-Ft|/Yt'] = zoo['|Yt-Ft|'] / zoo['ZOO Visitors']
zoo
```

```python
zoo.loc[:, '|Yt-Ft|'].mean()
```

```python
zoo.loc[:, '|Yt-Ft|/Yt'].mean()
```

```python
m= zoo['Yt-Ft']**2
m
```

```python
m.mean()#MSE
```

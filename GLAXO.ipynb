```python
GL_df=GL_df[['Date', 'Close']]
GL_df
```

```python
GL_df=GL_df.set_index(pd.DatetimeIndex(GL_df['Date']))
GL_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sn
%matplotlib inline
```

```python
GL_df.plot( y=['Close'])
plt.show()
```

```python
GL_df['gain']=GL_df.Close.pct_change(periods=1)
GL_df.head(5)
```

```python
GL_df=GL_df.dropna()
GL_df
```

```python
plt.figure(figsize=(8,6))
plt.plot(GL_df.index, GL_df.gain)
plt.xlabel=('Time')
plt.ylabel=('gain')
```

```python
round(GL_df.gain.mean(),4)
```

```python
round(GL_df.gain.std(),4)
```

```python
gain=GL_df
gain
```

```python
gain['forecasted_gain']=gain.gain.rolling(2).mean()
gain
```

```python
gain=gain[['gain', 'forecasted_gain']]
gain
```

```python
gain.mean()#MSE
```

```python
round(gain.gain.mean(),4)
```

```python
gain['Yt-Ft'] = gain['gain'] - gain['forecasted_gain']
gain
```

```python
gain.plot(y=['gain','forecasted_gain'])
plt.title("Gain Vs. Forecasted Gain")
plt.show()
```

```python
gain['|Yt-Ft|'] = gain['Yt-Ft'].abs()
gain
```

```python
#MAPE
gain['|Yt-Ft|/Yt'] = gain['|Yt-Ft|'] / gain['gain']
gain
```

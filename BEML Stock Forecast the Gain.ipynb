```python
gain=beml_df
```

```python
gain
```

```python
gain.describe()
```

```python
gain['forecasted_gain']=gain.gain.rolling(2).mean()
gain
```

```python
gain=gain[['gain', 'forecasted_gain']]
gain
```

```python
gain.mean()#MSE
```

```python
round(gain.gain.mean(),4)
```

```python
gain['Yt-Ft'] = gain['gain'] - gain['forecasted_gain']
gain
```

```python
gain.plot(y=['gain','forecasted_gain'])
plt.title("Gain Vs. Forecasted Gain")
plt.show()
```

```python
gain['|Yt-Ft|'] = gain['Yt-Ft'].abs()
gain
```

```python
gain['|Yt-Ft|/Yt'] = gain['|Yt-Ft|'] / gain['gain']
gain
#MAPE
```

```python
import pandas as pd
from sklearn import linear_model
import statsmodels.api as sm
```

```python
cook=pd.read_csv("C:\\Users\\HP\\Desktop\\Course\\Datas\\Cook.csv")
cook
```

```python
cook.info()
```

```python
cook.head(5)
```

```python
X=cook[['CompPrice','Income','Promotion','Price','Age',]]
X
```

```python
y=cook['Sales']
y
```

```python
# with sklearn
regr = linear_model.LinearRegression()
regr.fit(X, y)
```

```python
print('Intercept: \n', regr.intercept_)
print('Coefficients: \n', regr.coef_)
```

```python
x = sm.add_constant(X) # adding a constant
 
model = sm.OLS(y, x).fit()
predictions = model.predict(x) 
 
print_model = model.summary()
print(print_model)
```

```python
from sklearn.metrics import mean_absolute_percentage_error
from sklearn.model_selection import train_test_split
```

```python
mlrX=x
mlrX
```

```python
mlrY=cook['Sales']
mlrY
```

```python
MLRtrain_X, MLRtest_X, MLRtrain_Y, MLRtest_Y=train_test_split(mlrX,mlrY, train_size=0.8,random_state=42)
```

```python
MLRpredict_y=model.predict(MLRtest_X[MLRtrain_X.columns])
MLRpredict_y
```

```python
import numpy as np
```

```python
np.abs(mean_absolute_percentage_error(MLRtest_Y, MLRpredict_y))
```

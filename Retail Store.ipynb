```python
item_revenue = df.groupby('Total Revenue')['Total Revenue'].sum()
item_revenue
```

```python
# Group the data by 'Item' and calculate the sum of 'Sales' for each item
item_rice = df.groupby('Rice')['Total Revenue'].sum()
item_rice
```

```python
# Calculate the percentage of sales revenue for each item
percentage_rice = (item_rice / item_revenue.sum()) * 100
percentage_rice

```
